# ---Univariate Analysis---

### Making the Categories of "education" Column

```python
def clean_education(x):
    if x=="Bachelors": 
        return "Bachelor's Degree"
    if x=="Assoc-voc":
        return "Bachelor's Degree"
    if x=="Masters": 
        return "Master's Degree" 
    if x=="Prof-school":
         return "Master's Degree"
    if x=="Doctorate":
        return "Doctorate Degree"
    else:
        return "Less than Bachelor's"
    
```

```python
data["education"]=data["education"].apply(clean_education)
```

```python
data["education"].unique()
```
